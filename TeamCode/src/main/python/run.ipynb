```python
import cv2 
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(256, 256, 3)))
model.add(tf.keras.layers.MaxPooling2D((2, 2)))
model.add(tf.keras.layers.Conv2D(64, (3, 3), activation='relu'))
model.add(tf.keras.layers.MaxPooling2D((2, 2)))
model.add(tf.keras.layers.Conv2D(64, (3, 3), activation='relu'))
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(64, activation='selu'))
model.add(tf.keras.layers.Dropout(0.5))
model.add(tf.keras.layers.Dense(2, activation='softmax'))
```

```python
model.load_weights('model.weights.best_200.hdf5')
def predict(image):
    image = cv2.resize(image, (256, 256))
    image = image.reshape(1, 256, 256, 3)
    prediction = model.predict(image)
    return prediction[0]
```

```python
vid = cv2.VideoCapture(0) 
vid.set(cv2.CAP_PROP_FRAME_WIDTH, 1440)
vid.set(cv2.CAP_PROP_FRAME_HEIGHT, 960)
#     image_arr = np.array(img) 
#     # image_arr = cv2.GaussianBlur(image_arr, (101, 101), 0)
#     image_1 = image_arr[0:320,0:480] 
#     image_2 = image_arr[0:320,480:960]
#     image_3 = image_arr[0:320,960:1440]
#     image_1 = Image.fromarray(image_1)
#     image_2 = Image.fromarray(image_2)
#     image_3 = Image.fromarray(image_3)
while(True): 
    ret, frame = vid.read() 
    left = frame[140:290,220:420] 
    right = frame[140:290,820:1020]
    mid = frame[120:270,520:720]
    pre_left = predict(left)
    pre_right = predict(right)
    pre_mid = predict(mid)
    print(pre_left, pre_mid, pre_right)
    
    
    frame = cv2.rectangle(frame, (220,140), (420,290), (255,255,255), 3)
    frame = cv2.rectangle(frame, (820,140), (1020,290), (255,255,255), 3)
    frame = cv2.rectangle(frame, (520,120), (720,270), (255,255,255), 3)
    
    frame = cv2.putText(frame, str(pre_left), (50, 50), cv2.FONT_HERSHEY_SIMPLEX ,  
                   1, (255, 0, 0) , 2, cv2.LINE_AA) 
    frame = cv2.putText(frame, str(pre_mid), (50, 100), cv2.FONT_HERSHEY_SIMPLEX ,  
                   1, (255, 0, 0) , 2, cv2.LINE_AA)
    frame = cv2.putText(frame, str(pre_right), (50, 150), cv2.FONT_HERSHEY_SIMPLEX ,  
                   1, (255, 0, 0) , 2, cv2.LINE_AA)

    maxside = None
    if pre_left[1] > pre_mid[1] and pre_left[1]> pre_right[1] and pre_left[1] > 0.1:
        maxside = "left"
    if pre_mid[1] > pre_left[1] and pre_mid[1] > pre_right[1] and pre_mid[1] > 0.1:
        maxside = "mid"
    if pre_right[1] > pre_mid[1] and pre_right[1] > pre_left[1] and pre_right[1] > 0.1:
        maxside = "right"
    
    frame = cv2.putText(frame, str(maxside), (700, 50), cv2.FONT_HERSHEY_SIMPLEX ,  
                   1, (255, 0, 0) , 2, cv2.LINE_AA)
    
    cv2.imshow('frame', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
vid.release() 
cv2.destroyAllWindows() 
```
